```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/archive.zip")
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
#histogram for weight
plt.hist(data['Weight(Pounds)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
#visualizing the histograms
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 1, 1)
plt.hist(data['Height(Inches)'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
#visualizing the histograms
plt.tight_layout()
plt.show()
```
