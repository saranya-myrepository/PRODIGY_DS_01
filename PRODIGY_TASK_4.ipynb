```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = "/content/archive (5).zip"
social_media_data = pd.read_csv("/content/archive (5).zip")


common_sentiments = social_media_data['Sentiment'].value_counts().nlargest(5).index
filtered_data = social_media_data[social_media_data['Sentiment'].isin(common_sentiments)]


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_data, x='Sentiment', palette="viridis")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(12, 6))
sentiment_trends = filtered_data.groupby(['Year', 'Month', 'Sentiment']).size().reset_index(name='Count')
sns.lineplot(data=sentiment_trends, x='Month', y='Count', hue='Sentiment', style='Year', markers=True, palette="muted")
plt.title("Sentiment Trends Over Time (Top 5 Sentiments)")
plt.xlabel("Month")
plt.ylabel("Number of Posts")
plt.legend(title="Sentiment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_data, x='Platform', hue='Sentiment', palette="coolwarm")
plt.title("Platform-Based Sentiment Distribution")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.legend(title="Sentiment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


top_countries = social_media_data['Country'].value_counts().nlargest(5).index
country_filtered_data = filtered_data[filtered_data['Country'].isin(top_countries)]
plt.figure(figsize=(12, 6))
sns.countplot(data=country_filtered_data, x='Country', hue='Sentiment', palette="Set2")
plt.title("Country-Based Sentiment Distribution (Top 5 Countries)")
plt.xlabel("Country")
plt.ylabel("Count")
plt.legend(title="Sentiment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
